```python
questions=("1.Which year is this?","2.Which decade is this","3.Which century is this?")
options=(("A.2024","B.2016","C.2026","D.2025"),("A.1st","B.2nd","C.3rd","D.4th"),("A.21st","B.22nd","C.23rd","D.24th"))
answers=("B","C","A")
money=0
for i in range(len(questions)):
    print(questions[i])
    for opt in options[i]:
        print(opt)
    user=input("Which is the correct option:").upper()
    if(user==answers[i]):
      print("Correct")
      money=money+1000
    else:
      print('U lost better luck next time')
      break

print("U got this much of money $",money)
```
